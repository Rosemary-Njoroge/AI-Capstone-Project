Import statements

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Read the data

```python
data = pd.read_csv("Mall_Customers.csv")
data.head()
```

Check for nulls

```python
data.isna().any()
```

Check for duplicates

```python
data.duplicated().any()
```

Descriptive statistics

```python
data.shape
print(f"The data has {data.shape[0]} rows and {data.shape[1]} columns")
```

```python
data.info()
```

```python
data.describe()
```

Correlation between features

```python
numeric_cols = data.select_dtypes(["int64"])
correlation_matrix = numeric_cols.corr()
correlation_matrix["Spending Score (1-100)"].sort_values(ascending=False)
```

```python
sns.heatmap(correlation_matrix, annot=True)
```

Analysis of Spending Score

```python
sns.histplot(data["Spending Score (1-100)"], kde=True)
```

```python
sns.boxplot(data=data, x="Spending Score (1-100)")
```

Distribution of customer age

```python
plt.hist(data["Age"])
plt.show()
```

Relationship between customer age and spending score

```python
sns.scatterplot(data, x="Age", y="Spending Score (1-100)")
```

```python
print("Customers aged 40 and below have a higher spending score")
```

Distribution of Gender 

```python
gender_counts = data["Gender"].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.title("Customer gender distribution")
plt.xlabel("Gender")
plt.ylabel("Number of customers")
plt.show()
```

Relationship between gender and spending score

```python
sns.boxplot(data, x="Gender", y="Spending Score (1-100)")
```

Distribution of income among customers

```python
sns.histplot(data["Annual Income (k$)"], kde=True)
```

Relationshp between Income and spending score

```python
sns.scatterplot(data, x="Annual Income (k$)", y="Spending Score (1-100)")
plt.title("Income vs Spending Score")
plt.show()
```

```python
print("A customer's income doesn't seem to have much effect on their spending score")
```
